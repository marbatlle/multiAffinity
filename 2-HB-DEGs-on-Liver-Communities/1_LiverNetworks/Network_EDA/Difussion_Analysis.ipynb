```python
import pandas as pd
import numpy as np
from scipy import stats
%matplotlib inline
```

```python
degs_path = '/home/mar/Documents/TFM/GitHub/HB_PublicData/2-HB-DEGs-on-Liver-Communities/3_HB_DEG_in_Communities/HB_db_DEG.csv'
degs = pd.read_csv(degs_path, index_col=0)
degs_names = degs.index.tolist()
len(degs_names)
```

```python
mat_path = '/home/mar/Documents/TFM/GitHub/HB_PublicData/2-HB-DEGs-on-Liver-Communities/1_LiverNetworks/Network_EDA/PTmatrix_liver_5.txt'
mat = pd.read_csv(mat_path, sep='\t', index_col=0)
```

```python
mat.head()
```

```python
mat_degs = mat.filter(degs_names)
mat_degs = mat_degs[mat_degs.index.isin(degs_names)]
```

```python
result = pd.concat([degs, mat_degs], axis=1, join='inner')
```

```python
result
```

```python
print(abs(result['effect size']))
```

```python
genes = result.index

for i in genes:
    corr = abs(result['effect size']).corr(result[i])
    if corr > 0.2:
        print(i,': ',corr)
    elif corr < -0.2:
        print(i,': ',corr)
```
