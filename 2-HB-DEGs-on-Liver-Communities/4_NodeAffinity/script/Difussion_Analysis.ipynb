```python
import pandas as pd
import numpy as np
from scipy import stats
%matplotlib inline
```

```python
degs_path = '/home/mar/Documents/TFM/GitHub/HB_PublicData/1-Obtaining-DEGs-for-HB/DEGs_HB/HB_db_DEG.csv'
degs = pd.read_csv(degs_path, index_col=0)
degs_names = degs.index.tolist()
len(degs_names)
```

```python
mat_path = '/home/mar/Documents/TFM/GitHub/HB_PublicData/2-HB-DEGs-on-Liver-Communities/1_LiverNetworks/Network_EDA/PTmatrix_liver.txt'
mat = pd.read_csv(mat_path, sep='\t', index_col=0)
```

```python
mat_degs = mat.filter(degs_names)
mat_degs = mat_degs[mat_degs.index.isin(degs_names)]
mat_degs = mat_degs.rename_axis('gene')
```

```python
result = pd.merge(degs, mat_degs,on='gene', how='right')
result
```

```python
genes = result.index

for i in genes:
    corr = abs(result['effect size']).corr(result[i])
    if corr > 0.1:
        print(i,': ',corr)
    elif corr < -0.1:
        print(i,': ',corr)
```

```python
corr_list = []

for i in genes:
    corr = abs(result['effect size']).corr(result[i])
    corr_list.append(corr)

corr_list
```
